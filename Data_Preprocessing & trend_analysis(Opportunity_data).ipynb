**Handling Outliers and Anomalies**

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Convert date columns to datetime format
date_columns = ["Opportunity End Date", "Apply Date", "Opportunity Start Date"]
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Handling missing values
# Create a dictionary to specify the columns and their fill values
fill_values = {
    "Graduation Date(YYYY MM)": df["Graduation Date(YYYY MM)"].mode()[0],
    "Current Student Status": df["Current Student Status"].mode()[0],
    "Current/Intended Major": "Unknown",
    "Opportunity Start Date": df["Opportunity Start Date"].mode()[0]
}

# Use fillna with the dictionary to perform the operation on the original DataFrame
df.fillna(value=fill_values, inplace=True)
```

**Normalize numerical features**

```python
# Normalize numerical features
scaler = MinMaxScaler()
df[["Reward Amount", "Skill Points Earned"]] = scaler.fit_transform(df[["Reward Amount", "Skill Points Earned"]])
```

**One-hot encoding categorical variables**

```python
# One-hot encoding categorical variables
categorical_cols = ["Opportunity Category", "Status Description", "Current Student Status"]
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Compute opportunity duration (in days)
df['Opportunity Duration'] = (df['Opportunity End Date'] - df['Opportunity Start Date']).dt.days
```

```python
# Dropping colums
df.drop(columns=['Profile Id', 'Opportunity Id', 'Zip Code'], inplace=True)
```

**Key Insights**

The dataset has been successfully preprocessed with the following improvements:

Dates Converted: Opportunity End Date, Apply Date, and Opportunity Start Date are now in datetime format.

Missing Values Handled: Key columns like Graduation Date, Current Student Status, and Current/Intended Major are filled with appropriate values.

Normalization Applied: Reward Amount and Skill Points Earned are scaled using Min-Max normalization.

Categorical Encoding: One-hot encoding applied to Opportunity Category, Status Description, and Current Student Status.

Feature Engineering: Opportunity Duration (days between start and end date) is calculated.

Dropped Unnecessary Columns: Profile Id, Opportunity Id, and Zip Code are removed.

**TREND ANALYSIS**

```python
# Extracting 'year' and 'month' from the relevant date columns
df["Apply Month"] = df["Apply Date"].dt.to_period("M")
df["Start Month"] = df["Opportunity Start Date"].dt.to_period("M")
df["End Month"] = df["Opportunity End Date"].dt.to_period("M")

# Aggregate data by month (number of applications, opportunities started, and ended)
apply_trend = df["Apply Month"].value_counts().sort_index()
start_trend = df["Start Month"].value_counts().sort_index()
end_trend = df["End Month"].value_counts().sort_index()

# Plot the trend analysis
plt.figure(figsize=(12, 6))
plt.plot(apply_trend.index.astype(str), apply_trend.values, marker="o", label="Applications")
plt.plot(start_trend.index.astype(str), start_trend.values, marker="s", label="Opportunities Started")
plt.plot(end_trend.index.astype(str), end_trend.values, marker="^", label="Opportunities Ended")

plt.xticks(rotation=45)
plt.xlabel("Month-Year")
plt.ylabel("Count")
plt.title("Trend Analysis of Applications, Opportunities Started, and Ended")
plt.legend()
plt.grid()
plt.show()
```

```python
# 1. Comparative Analysis: Average Reward Amount by Opportunity Category
plt.figure(figsize=(10, 5))
# Get the one-hot encoded columns for 'Opportunity Category'
opportunity_category_cols = [col for col in df.columns if col.startswith("Opportunity Category_")]
# Melt the DataFrame to combine these columns into a single 'Opportunity Category' column
df_melted = df.melt(id_vars=["Reward Amount"], value_vars=opportunity_category_cols, var_name="Opportunity Category", value_name="Category Value")
df_melted = df_melted[df_melted["Category Value"] == 1]  # Keep only rows where category is active
# Plot the barplot using the melted DataFrame
sns.barplot(x="Opportunity Category", y="Reward Amount", data=df_melted, estimator=lambda x: sum(x)/len(x))
plt.xticks(rotation=45)
plt.title("Average Reward Amount by Opportunity Category")
plt.xlabel("Opportunity Category")
plt.ylabel("Average Reward Amount")
plt.show()
```

```python
# 2. Comparative Analysis: Skill Points Earned by Student Status
plt.figure(figsize=(10, 5))
student_status_cols = [col for col in df.columns if "Current Student Status_" in col]
df_melted = df.melt(id_vars=["Skill Points Earned"], value_vars=student_status_cols, var_name="Student Status", value_name="Status Value")
df_melted = df_melted[df_melted["Status Value"] == 1]  # Keep only rows where status is active

sns.barplot(x="Student Status", y="Skill Points Earned", data=df_melted, estimator=lambda x: sum(x)/len(x))
plt.xticks(rotation=45)
plt.title("Skill Points Earned by Student Status")
plt.xlabel("Student Status")
plt.ylabel("Average Skill Points Earned")
plt.show()
```

```python
# 3. Comparative Analysis: Distribution of Status Description across Countries
plt.figure(figsize=(12, 6))
status_cols = [col for col in df.columns if "Status Description_" in col]
df_status = df.groupby("Country")[status_cols].sum()
df_status.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Distribution of Status Description across Countries")
plt.xlabel("Country")
plt.ylabel("Count of Statuses")
plt.legend(title="Status Description", bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
# Loading dataset
df = pd.read_csv('/content/user_cleaned_data.csv')
df.head()
```

```python
# Convert 'Sign Up Date' to datetime format
df['Sign Up Date'] = pd.to_datetime(df['Sign Up Date'], errors='coerce')

# Standardize categorical columns: Replace 'Unknown' and 'Invalid' with NaN
df.replace({'Unknown': None, 'Invalid': None}, inplace=True)

# Extract useful time-based features from 'Sign Up Date'
df['Sign Up Year'] = df['Sign Up Date'].dt.year
df['Sign Up Month'] = df['Sign Up Date'].dt.month

# Drop 'Sign Up Date' after extracting relevant features
df.drop(columns=['Sign Up Date'], inplace=True)

# Convert boolean column to integer (0 and 1)
df['isFromSocialMedia'] = df['isFromSocialMedia'].astype(int)

# Display the cleaned dataset info
df.info(), df.head()
```

```python
# Replace 'Unknown' or 'Invalid' with NaN for proper handling
df.replace({'Unknown': None, 'Invalid': None}, inplace=True)

# Fill missing categorical values with the mode of each column
# Remove 'Gender' and 'Degree' from categorical_cols as they are already one-hot encoded
categorical_cols = ['city', 'zip']
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

df['isFromSocialMedia'] = df['isFromSocialMedia'].astype(int)

# Display final dataset info after preprocessing
df.info(), df.head()
```

```python
# TREND ANALYSIS

import matplotlib.pyplot as plt

# Aggregate user sign-ups by year
trend_data = df.groupby(['Sign Up Year'])['isFromSocialMedia'].count()

# Plot trend over time
plt.figure(figsize=(10, 5))
plt.plot(trend_data.index, trend_data.values, marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Number of Users")
plt.title("User Sign-Ups Trend Over Time")
plt.grid(True)
plt.show()
```

```python
# Assuming 'Gender_Female', 'Gender_Male', and potentially 'Gender_Other' are columns in your DataFrame 'df'

# Create a new 'Gender' column based on the existing gender columns
df['Gender'] = 'Gender_Other'  # Initialize with 'Gender_Other' as default
df.loc[df['Gender_Female'] == 1, 'Gender'] = 'Gender_Female'  # Set to 'Gender_Female' if Gender_Female is 1
df.loc[df['Gender_Male'] == 1, 'Gender'] = 'Gender_Male'  # Set to 'Gender_Male' if Gender_Male is 1


# Drop the original gender columns
df.drop(columns=['Gender_Female', 'Gender_Male'], inplace=True)

print(df['Gender'].value_counts())
```

```python
# COMPARATIVE ANALYSIS

import seaborn as sns

# Check if 'Gender' column exists in the DataFrame
if 'Gender' in df.columns:
    # Compare user sign-ups by Gender
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='Sign Up Year', hue='Gender')
    plt.xlabel("Year")
    plt.ylabel("Number of Users")
    plt.title("User Sign-Ups by Gender Over Time")
    plt.legend(title="Gender")
    plt.show()
else:
    print("Gender column not found in the DataFrame. Please check your data.")
```
