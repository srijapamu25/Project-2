**Loading Libraries**

```python
import pandas as pd
import numpy as np
```

**Loading dataset**

```python
df = pd.read_csv('/content/Week-1/UserData.csv')
df
```

**DATA CLEANING**

```python
# checking missing values
df.isnull().sum()
```

```python
# duplicate rows
df.duplicated().sum()
```

**Handling Missing values**

```python
# Filling missing values with 'Unknown' for 'Gender', 'Degree', 'city' and 'N/A' for 'zip'

df.fillna({"Gender": "Unknown", "Degree": "Unknown", "city": "Unknown", "zip": "N/A"}, inplace=True)
df.fillna({'Country': 'Unknown'}, inplace=True)
df.fillna({'isFromSocialMedia': 'False'}, inplace=True)
df
```

```python
# Checking and dropping duplicates
df.drop_duplicates(inplace=True)
df
```

**Standardizing formats**

```python
# Converting 'Sign Up Date' to datetime format
df['Sign Up Date'] = pd.to_datetime(df['Sign Up Date'])

# Standardize city names
df['city'] = df['city'].str.title()

# Standardize country names
df['Country'] = df['Country'].str.title()

df[['Sign Up Date', 'city', 'Country']].head()
```

**DATA VALIDATION**

```python
# Checking unique values in continuous variable
df["zip"].unique()
df["zip"] = df["zip"].apply(lambda x: x if x.isnumeric() else "Invalid")
print(df["zip"].unique()[:10])
```

```python
# Checking unique values in categorical variable
categorical_columns = ['Gender', 'Country', 'Degree', 'city', 'isFromSocialMedia']
for col in categorical_columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python
# Cross checking
df['PreferredSponsors'] = df.apply(
    lambda row: "Unknown" if row["isFromSocialMedia"] == "True" and row["PreferredSponsors"] == "[]" else row["PreferredSponsors"],
    axis=1
)
```

```python
# saving the cleaned dataset
df.to_csv('cleaned_data.csv', index=False)
```
