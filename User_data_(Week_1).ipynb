**Loading Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading dataset**

```python
df = pd.read_csv('/content/Week-1/UserData.csv')
df
```

```python
df.describe()
```

```python
# Shape
df.shape
```

```python
# Checking data types
df.dtypes
```

**DATA CLEANING**

```python
# checking missing values
df.isnull().sum()
```

```python
# duplicate rows
df.duplicated().sum()
```

**Handling Missing values**

```python
# Filling missing values with 'Unknown' for 'Gender', 'Degree', 'city' and 'N/A' for 'zip'

df.fillna({"Gender": "Unknown", "Degree": "Unknown", "city": "Unknown", "zip": "N/A"}, inplace=True)
df.fillna({'Country': 'Unknown'}, inplace=True)
df.fillna({'isFromSocialMedia': 'False'}, inplace=True)
df
```

```python
# Checking and dropping duplicates
df.drop_duplicates(inplace=True)
df
```

**Standardizing formats**

```python
# Converting 'Sign Up Date' to datetime format
df['Sign Up Date'] = pd.to_datetime(df['Sign Up Date'])

# Standardize city names
df['city'] = df['city'].str.title()

# Standardize country names
df['Country'] = df['Country'].str.title()

df[['Sign Up Date', 'city', 'Country']].head()
```

**DATA VALIDATION**

```python
# Checking unique values in continuous variable
df["zip"].unique()
df["zip"] = df["zip"].apply(lambda x: x if x.isnumeric() else "Invalid")
print(df["zip"].unique()[:10])
```

```python
# Checking unique values in categorical variable
categorical_columns = ['Gender', 'Country', 'Degree', 'city', 'isFromSocialMedia']
for col in categorical_columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python
# Cross checking
df['PreferredSponsors'] = df.apply(
    lambda row: "Unknown" if row["isFromSocialMedia"] == "True" and row["PreferredSponsors"] == "[]" else row["PreferredSponsors"],
    axis=1
)
```

```python
# saving the cleaned dataset
df.to_csv('cleaned_data.csv', index=False)
```

```python
# Summary
summary = {
    "Missing Values Handled": True,
    "Duplicates Removed": True,
    "Formats Standardized": True,
    "Numeric Data Validated": True,
    "Categorical Data Validated": True,
    "Relationships Cross-Checked": True
}

print(summary)
```

**EXPLORATORY DATA ANALYSIS**

```python
# Distribution of users by Gender

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Users by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Count occurrences of each degree
degree_counts = df["Degree"].value_counts().head(10)

# Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(degree_counts, labels=degree_counts.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Top 10 Degree Fields")
plt.axis("equal")
plt.show()
```

```python
# Density plot for Signup trends over time

plt.figure(figsize=(10, 5))
sns.kdeplot(df["Sign Up Date"].dt.year, shade=True, color="blue")
plt.title("Density of Signups Over Time")
plt.xlabel("Signup Year")
plt.ylabel("Density")
plt.show()
```

```python
# Bar graph for top 10 cities with most users

plt.figure(figsize=(10, 5))
top_cities = df["city"].value_counts().head(10)
sns.barplot(y=top_cities.index, x=top_cities.values, palette="coolwarm")
plt.title("Top 10 Cities with Most Users")
plt.xlabel("Number of Users")
plt.ylabel("city")
plt.show()
```

```python
# countries with similar backgrounds

plt.figure(figsize=(12, 6))
top_countries = df["Country"].value_counts().head(5).index  # Top 5 countries
filtered_df = df[df["Country"].isin(top_countries)]

sns.countplot(y=filtered_df["Degree"], hue=filtered_df["Country"], order=filtered_df["Degree"].value_counts().index[:10], palette="muted")
plt.title("Top Degrees by Country")
plt.xlabel("Count")
plt.ylabel("Degree Field")
plt.legend(title="Country")
plt.show()
```

```python
# Gender vs Degree

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Degree'].value_counts()
    for x_label, grp in df.groupby('Gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Gender')
_ = plt.ylabel('Degree')
```
